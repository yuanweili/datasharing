```python
def print_first_n_lines(text: str, *, n: int):
    print("\n".join(text.split("\n")[:n]), "...", sep="\n")
```

```python
import sec_parser as sp


elements: list = sp.Edgar10QParser().parse(html)

demo_output: str = sp.render(elements)
print_first_n_lines(demo_output, n=100)
```

```python
def show(elements):
    for element in elements:
        print(f"======================")
        text = element.text[:1000]
        if hasattr(element, "inner_elements"):
            print(f"->{element} (has {len(element.inner_elements)} elements inside)")
        elif text:
            print(f"{element} (text: {text}...)")
        else:
            print(f"{element}")
```

```python
show(elements)
```

```python
def showRiskText(elements):
    riskSection = False
    
    for element in elements:        
        if 'TitleElemen' in str(element):
            riskSection = False            
        
        if 'TitleElemen' in str(element) and 'Risk' in element.text:            
            riskSection = True
            print(f"--------------------")
            print('### ' + element.text)
            print(f"--------------------")
        
        if 'TitleElemen' not in str(element) and riskSection:
            print(element.text)

            
```

```python
 showRiskText(elements)
```

```python
def showCompanySummaryText(elements):
    companySummarySection = False
    
    for element in elements:        
        if 'TopSectionTitle' in str(element):
            companySummarySection = False            
        
        if 'TopSectionTitle' in str(element) and 'ITEM 1.' in element.text.upper():            
            companySummarySection = True
            print(f"--------------------")
            print('### ' + element.text)
            print(f"--------------------")

        if 'TopSectionTitle' in str(element) and ' BUSINESS' in element.text.upper():            
            companySummarySection = True
            print(f"--------------------")
            print('### ' + element.text)
            print(f"--------------------")
            
        if 'TitleElemen' in str(element) and 'ITEM 1.' in element.text.upper():            
            companySummarySection = True
            print(f"--------------------")
            print('### ' + element.text)
            print(f"--------------------")
            
            
        if 'TitleElemen' in str(element) and 'ITEM 1A.' in element.text.upper():
            companySummarySection = False

        if 'TitleElemen' in str(element) and 'RISK ' in element.text.upper():
            companySummarySection = False            
        
        if 'TitleElemen' in str(element) and companySummarySection:
            print(f"--------------------")
            print('#### ' + element.text)
            print(f"--------------------")
        
        if 'TitleElemen' not in str(element) and companySummarySection:            
            print('##### ' + element.text)

            
            

```

```python
showCompanySummaryText(elements)
```
